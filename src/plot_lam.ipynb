```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 创建数据集，调整顺序为SASRec+, ADRec, DiffuRec
data = {
    'Model': ['SASRec+', 'ADRec', 'DiffuRec'],
    'Baby': [67, 150, 741],
    'Beauty': [74, 173, 704],
    'ML-100K': [16, 27, 413],
    'Sports': [138, 304, 1140],
    'Toys': [316, 439, 809],
    'Yelp': [879, 2499, 5992]
}

# 转换为DataFrame
df = pd.DataFrame(data)
sns.set_context("paper")
# 使用melt函数将数据从宽格式转为长格式
df_melted = df.melt(id_vars=["Model"], var_name="Dataset", value_name="Train Time")

# 设置绘图风格
sns.set(style="whitegrid")
plt.figure(figsize=(6, 3))
colors = sns.color_palette("Set2", 3)
# 绘制柱状图
ax = sns.barplot(x="Dataset", y="Train Time", hue="Model", data=df_melted,palette=colors)
ax.bar_label(ax.containers[0], fontsize=8)
ax.bar_label(ax.containers[1], fontsize=8)
ax.bar_label(ax.containers[2], fontsize=8)
# plt.yscale('log')
# 设置标题和标签
# plt.title("Training Time for Different Models on Various Datasets", fontsize=14)
plt.xlabel("")
plt.ylabel("Training Time ($10^3$ s)")
ax.set_yticks([0, 1000, 2000, 3000, 4000, 5000,6000])  # 设置特定的x轴刻度
ax.set_yticklabels([0,1,2,3,4,5,6])  # 设置显示的刻度标签
# 显示图表
plt.tight_layout()
plt.savefig('../imgs/train_time.svg',bbox_inches='tight')
plt.show()
plt.close()
```
