```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.ticker as ticker

# 创建数据
data = {
    'Diffusion steps': [2, 4, 8, 16, 32, 64],
    'HR@20_Baby': [7.0502, 7.1081, 7.1519, 6.961, 7.0757, 7.0677],
    'HR@20_Beauty': [16.0696, 16.2548, 16.1398, 16.0975, 16.3833, 16.1889],
    'HR@20_ML-100K': [22.4408, 22.5778, 22.4605, 22.1666, 22.8126, 21.5412],
    'HR@20_Sports': [8.0827, 7.978, 7.9325, 8.4556, 8.3698, 8.3113],
    'HR@20_Toys': [11.8801, 11.7527, 12.4406, 12.3556, 12.3217, 12.1433],
    'HR@20_Yelp': [7.2284, 7.3057, 7.2877, 7.3326, 7.3976, 7.2789],
    'NDCG@20_Baby': [3.0887, 3.0425, 3.093, 3.022, 3.0985, 3.057],
    'NDCG@20_Beauty': [8.3223, 8.3159, 8.302, 8.3288, 8.433, 8.3859],
    'NDCG@20_ML-100K': [8.5192, 8.3746, 8.4417, 8.34, 8.4937, 8.0218],
    'NDCG@20_Sports': [3.5696, 3.6254, 3.5956, 3.7093, 3.6977, 3.6872],
    'NDCG@20_Toys': [6.6847, 6.6178, 6.8216, 6.8155, 6.8039, 6.7653],
    'NDCG@20_Yelp': [2.9225, 2.9512, 2.9504, 2.9753, 2.9964, 2.952]
}

# 将数据转换为 DataFrame
df = pd.DataFrame(data)

sns.set_context("paper")
# 设置绘图样式
sns.set(style="whitegrid", rc={"axes.grid": True, "grid.linestyle": "--", "grid.alpha": 0.5})

# 去除竖线网格线
# plt.grid(axis='x', linestyle='None')

# 创建子图
fig, axes = plt.subplots(1, 2, figsize=(7, 3.5))

# 绘制HR@20折线图
sns.lineplot(x='Diffusion steps', y='HR@20_Baby', data=df, marker='o', label='Baby', ax=axes[0])
sns.lineplot(x='Diffusion steps', y='HR@20_Beauty', data=df, marker='o', label='Beauty', ax=axes[0])
sns.lineplot(x='Diffusion steps', y='HR@20_ML-100K', data=df, marker='o', label='ML-100K', ax=axes[0])
sns.lineplot(x='Diffusion steps', y='HR@20_Sports', data=df, marker='o', label='Sports', ax=axes[0])
sns.lineplot(x='Diffusion steps', y='HR@20_Toys', data=df, marker='o', label='Toys', ax=axes[0])
sns.lineplot(x='Diffusion steps', y='HR@20_Yelp', data=df, marker='o', label='Yelp', ax=axes[0])
axes[0].set_title('')
axes[0].set_xlabel('Diffusion steps')
axes[0].set_ylabel('HR@20')
axes[0].legend([], [], frameon=False)
# 设置x轴为对数坐标并显示指定刻度
axes[0].set_xscale('log')
axes[0].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x)}'))
axes[0].set_xticks([2, 4, 8, 16, 32, 64])  # 设置特定的x轴刻度
axes[0].set_xticklabels([2, 4, 8, 16, 32, 64])  # 设置显示的刻度标签
axes[0].xaxis.set_minor_locator(ticker.LogLocator(base=2.0, subs='auto'))

# 绘制NDCG@20折线图
sns.lineplot(x='Diffusion steps', y='NDCG@20_Baby', data=df, marker='o', label='Baby', ax=axes[1])
sns.lineplot(x='Diffusion steps', y='NDCG@20_Beauty', data=df, marker='o', label='Beauty', ax=axes[1])
sns.lineplot(x='Diffusion steps', y='NDCG@20_ML-100K', data=df, marker='o', label='ML-100K', ax=axes[1])
sns.lineplot(x='Diffusion steps', y='NDCG@20_Sports', data=df, marker='o', label='Sports', ax=axes[1])
sns.lineplot(x='Diffusion steps', y='NDCG@20_Toys', data=df, marker='o', label='Toys', ax=axes[1])
sns.lineplot(x='Diffusion steps', y='NDCG@20_Yelp', data=df, marker='o', label='Yelp', ax=axes[1])
axes[1].set_title('')
axes[1].set_xlabel('Diffusion steps')
axes[1].set_ylabel('NDCG@20')

# 设置x轴为对数坐标并显示指定刻度
axes[1].set_xscale('log')
axes[1].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x)}'))
axes[1].set_xticks([2, 4, 8, 16, 32, 64])  # 设置特定的x轴刻度
axes[1].set_xticklabels([2, 4, 8, 16, 32, 64])  # 设置显示的刻度标签
axes[1].xaxis.set_minor_locator(ticker.LogLocator(base=2.0, subs='auto'))
axes[1].legend([], [], frameon=False)
# 获取两个子图中的所有线条对象
lines = axes[0].get_lines() 

# 创建一个共享的legend，位于上方
fig.legend(lines, [line.get_label() for line in lines], loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=6, frameon=False)

# 调整布局
plt.tight_layout()
# plt.savefig('../imgs/dif_steps.pdf',bbox_inches='tight')
plt.show()

plt.close()
```
